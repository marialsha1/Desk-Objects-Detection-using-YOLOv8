```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
from PIL import Image
import matplotlib.pyplot as plt

results = model("images", save = True)

for r in results:
    img = r.plot()
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
names = model.names

for r in results:
    print("Image: ", r.path)
    
    count = 0 #No. of objects detected
    
    for box in r.boxes:
        class_id = int(box.cls)
        class_names = names[class_id]
        confidence = float(box.conf)
        
        #To filter the results:
        if confidence > 0.6:
            print(f"Detected: {class_names} ({confidence: .2f} )")
            count += 1
    
    print(f"Total objects detected (filtered results): {count}")
    print("-----------")
```
